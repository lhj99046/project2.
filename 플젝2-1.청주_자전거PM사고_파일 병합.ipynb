```python
# 발생일시 -> datetime 으로 변환 -> date 컬럼 생성후 발생일시 컬럼 삭제
# 날짜 변환
df['year'] = df['발생일시'].astype(str)
```

```python
year = []
for i in range(len(df['year'])):
    a = df['year'][i][0:4]
    year.append(a)
df['year'] = year
```

```python
df['year']
```

```python
#위와 동일한 원리로 생성
df['month'] = df['발생일시'].astype(str)
month = []
for i in range(len(df['month'])):
    a = df['month'][i][4:6]
    month.append(a)
df['month'] = month
```

```python
df['day'] = df['발생일시'].astype(str)
day = []
for i in range(len(df['day'])):
    a = df['day'][i][6:8]
    day.append(a)
df['day'] = day
```

```python
df['hour'] = df['발생일시'].astype(str)
hour = []
for i in range(len(df['hour'])):
    a = df['hour'][i][-2:]
    hour.append(a)
df['hour'] = hour
```

```python
df['date'] = df['year'] + '-' + df['month'] + '-' + df['day'] + ' ' + df['hour'] + ':00:00'
df['date'] = pd.to_datetime(df['date']) #datetime 으로 변환
```

```python
df.drop('발생일시', axis = 1, inplace = True) #axis = 0 : 행 / axis = 1 : 열 /inplace=True : 원본에서 바로 바꾸기.
```
